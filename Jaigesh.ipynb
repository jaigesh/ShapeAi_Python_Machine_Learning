```python
import numpy as np
import pandas as pd
import sklearn as sl
from sklearn.datasets import load_boston
df = load_boston()
boston = pd.DataFrame(df.data, columns=df.feature_names)
boston.head()
boston['Target'] = df.target
boston.head()
boston.isnull().sum()
from sklearn.model_selection import train_test_split
X = boston.drop('Target', axis=1)
Y = boston['Target']
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.15, random_state = 5)
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
lin_model = LinearRegression()
lin_model.fit(X_train, Y_train)
y_train_predict = lin_model.predict(X_train)
MSE = (np.sqrt(mean_squared_error(Y_train, y_train_predict)))
print('The model preformance for training set')
print('MSE is {} '.format(MSE))
print('\n')
y_test_predict = lin_model.predict(X_test)
MSE = (np.sqrt(mean_squared_error(Y_test, y_test_predict)))
print('The model preformance for testing set')
print('MSE is {}'.format(MSE))
```
